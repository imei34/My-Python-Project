```python
#import library yang dibutuhkan yaitu numpy
import numpy as np
```

```python
#Gauss Seidel 3x3 dengan menggunakan python
#deklarasi variabel

#C sebagai matriks variabel x1,x2,x3
C = np.array([[12,3,-5],
             [1,5,3],
             [3,7,13]])

#X sebagai matriks hasil x1,x2,x3
X = np.array([[1],
                [28],
                [76]])

#XG sebagai matriks tebakan
XG = np.array([[1],
              [0],
              [1]])
```

```python
#A sebagai matriks variabel x1,x2,x3
#x sebagai matriks hasil x1,x2,x3
#xg sebagai matriks tebakan

def gaussSeidel3x3(A,x,xg):
    print("-"*10)
    print("-"*10)
    print("iterasi ke-1")
    xnew1 = np.array((x[0] -A[0][1]*xg[1] -A[0][2]*xg[2])/A[0][0])
    print(f"x1 = {float(x[0])} - {A[0][1]}x2 - {A[0][2]}x3 / {A[0][0]}")
    print(f"x1 = {float(x[0])} - {A[0][1]}({float(xg[1])}) - {A[0][2]}({float(xg[2])}) / {A[0][0]}")
    print(f"x1 = {float(np.array((x[0] -A[0][1]*xg[1] -A[0][2]*xg[2])))}/{A[0][0]}")
    print(f"x1 = {float(np.array((x[0] -A[0][1]*xg[1] -A[0][2]*xg[2])/A[0][0]))}\n")
    
    
    xnew2 = np.array((x[1] -A[1][0]*xnew1 -A[1][2]*xg[2])/A[1][1])
    print(f"x2 = {float(x[1])} - {A[1][0]}x1 - {A[1][2]}x3 / {A[1][1]}")
    print(f"x2 = {float(x[1])} - {A[1][0]}({float(xnew1)}) - {A[1][2]}({float(xg[2])}) / {A[1][1]}")
    print(f"x2 = {float(np.array((x[1] -A[1][0]*xnew1 -A[1][2]*xg[2])))}/{A[1][1]}")
    print(f"x2 = {float(np.array((x[1] -A[1][0]*xnew1 -A[1][2]*xg[2])/A[1][1]))}\n")
    
    xnew3 = np.array((x[2] -A[2][0]*xnew1 -A[2][1]*xnew2)/A[2][2])
    print(f"x3 = {float(x[2])} - {A[2][0]}x1 - {A[2][1]}x2 / {A[2][2]}")
    print(f"x3 = {float(x[2])} - {A[2][0]}({float(xnew1)}) - {A[2][1]}({float(xnew2)}) / {A[2][2]}")
    print(f"x3 = {float(np.array((x[2] -A[2][0]*xnew1 -A[2][1]*xnew2)))}/{A[2][2]}")
    print(f"x3 = {float(np.array((x[2] -A[2][0]*xnew1 -A[2][1]*xnew2)/A[2][2]))}\n")
    
    err1 = np.abs((xnew1-xg[0])*100/xnew1)
    print(f"error 1 = |{float(np.abs(xnew1))}-{float(np.abs(xg[0]))}/{float(np.abs(xnew1))}|*100")
    print(f"error 1 = |{float(np.abs(xnew1-xg[0]))}/{float(np.abs(xnew1))}|*100")
    print(f"error 1 = {float(np.abs((xnew1-xg[0])*100/xnew1))}%\n")
    
    err2 = np.abs((xnew2-xg[1])*100/xnew2)
    print(f"error 2 = |{float(np.abs(xnew2))}-{float(np.abs(xg[1]))}/{float(np.abs(xnew2))}|*100")
    print(f"error 2 = |{float(np.abs(xnew2-xg[1]))}/{float(np.abs(xnew2))}|*100")
    print(f"error 2 = {float(np.abs((xnew2-xg[1])*100/xnew2))}%\n")
    
    err3 = np.abs((xnew3-xg[2])*100/xnew3)
    print(f"error 3 = |{float(np.abs(xnew3))}-{float(np.abs(xg[2]))}/{float(np.abs(xnew3))}|*100")
    print(f"error 3 = |{float(np.abs(xnew3-xg[2]))}/{float(np.abs(xnew3))}|*100")
    print(f"error 3 = {float(np.abs((xnew3-xg[2])*100/xnew3))}%\n")
    
    print(np.array([[xnew1],
                       [xnew2],
                       [xnew3]]))
    
    print("\n")
    print(f"error 1 = {float(err1)}%")
    print(f"error 2 = {float(err2)}%")
    print(f"error 3 = {float(err3)}%")
    print("-"*10)
    print("-"*10)
    
    #atur loop (start,end) untuk jumlah iterasi gauss seidell
    for i in range(2,8):
        print("-"*10)
        print("-"*10)
        print("iterasi ke-",i)
        xold1 = xnew1
        xnew1 = np.array((x[0] -A[0][1]*xnew2 -A[0][2]*xnew3)/A[0][0])
        print(f"x1 = {float(x[0])} - {A[0][1]}x2 - {A[0][2]}x3 / {A[0][0]}")
        print(f"x1 = {float(x[0])} - {A[0][1]}({float(xnew2)}) - {A[0][2]}({float(xnew3)}) / {A[0][0]}")
        print(f"x1 = {float(np.array((x[0] -A[0][1]*xnew2 -A[0][2]*xnew3)))}/{A[0][0]}")
        print(f"x1 = {float(np.array((x[0] -A[0][1]*xnew2 -A[0][2]*xnew3)/A[0][0]))}\n")
        
        xold2 = xnew2
        xnew2 = np.array((x[1] -A[1][0]*xnew1 -A[1][2]*xnew3)/A[1][1])
        print(f"x2 = {float(x[1])} - {A[1][0]}x1 - {A[1][2]}x3 / {A[1][1]}")
        print(f"x2 = {float(x[1])} - {A[1][0]}({float(xnew1)}) - {A[0][2]}({float(xnew3)}) / {A[1][1]}")
        print(f"x2 = {float(np.array((x[1] -A[1][0]*xnew1 -A[1][2]*xnew3)))}/{A[1][1]}")
        print(f"x2 = {float(np.array((x[1] -A[1][0]*xnew1 -A[1][2]*xnew3)/A[1][1]))}\n")
        
        xold3 = xnew3
        xnew3 = np.array((x[2] -A[2][0]*xnew1 -A[2][1]*xnew2)/A[2][2])
        print(f"x3 = {float(x[2])} - {A[2][0]}x1 - {A[2][1]}x2 / {A[2][2]}")
        print(f"x3 = {float(x[2])} - {A[2][0]}({float(xnew1)}) - {A[2][1]}({float(xnew2)}) / {A[2][2]}")
        print(f"x3 = {float(np.array((x[2] -A[2][0]*xnew1 -A[2][1]*xnew2)))}/{A[2][2]}")
        print(f"x3 = {float(np.array((x[2] -A[2][0]*xnew1 -A[2][1]*xnew2)/A[2][2]))}\n")
        
        #err1,err2,err3 berfungsi untuk paramater error nilai x1,x2,x3
        #semakin kecil nilai error maka semakin baik
        #nilai error yang baik adalah dibawah 1%
        err1 = float(np.abs((xnew1-xold1)*100/xnew1))
        print(f"error 1 = |{float(np.abs(xnew1))}-{float(np.abs(xold1))}/{float(np.abs(xnew1))}|*100")
        print(f"error 1 = |{float(np.abs(xnew1-xold1))}/{float(np.abs(xnew1))}|*100")
        print(f"error 1 = {float(np.abs((xnew1-xold1)*100/xnew1))}%\n")
        
        err2 = float(np.abs((xnew2-xold2)*100/xnew2))
        print(f"error 1 = |{float(np.abs(xnew2))}-{float(np.abs(xold2))}/{float(np.abs(xnew2))}|*100")
        print(f"error 1 = |{float(np.abs(xnew2-xold2))}/{float(np.abs(xnew2))}|*100")
        print(f"error 1 = {float(np.abs((xnew2-xold2)*100/xnew2))}%\n")
        
        err3 = float(np.abs((xnew3-xold3)*100/xnew3))
        print(f"error 1 = |{float(np.abs(xnew3))}-{float(np.abs(xold3))}/{float(np.abs(xnew3))}|*100")
        print(f"error 1 = |{float(np.abs(xnew3-xold3))}/{float(np.abs(xnew3))}|*100")
        print(f"error 1 = {float(np.abs((xnew3-xold3)*100/xnew3))}%\n")
        
        print(np.array([[xnew1],
                       [xnew2],
                       [xnew3]]))
        print("\n")
        print(f"error 1 = {float(err1)}%")
        print(f"error 2 = {float(err2)}%")
        print(f"error 3 = {float(err3)}%")
        print("-"*10)
        print("-"*10)
```

```python
gaussSeidel3x3(C,X,XG)
```
